## **Quando Usar e Como Plotar um Gráfico de Waffle**

### **Resumo**

Uma das atividades mais comuns para o cientista de dados é a criação de gráficos para explanação dos seus resultados e busca por *insigths*. Logo, um elemento muito importante do *Data Science* é a visualização de dados, sendo uma das coisas que, particularmente, mais gosto em projetos de ciência de dados.

Diferentemente das tabelas que interagem como nosso sistema verbal, os gráficos interagem com nosso sistema visual, que processa as informações com mais rapidez. Assim, um gráfico bem desenhado certamente comunicará a informação com mais rapidez que uma tabela bem desenhada.

Existem vários tipos de gráfico, linhas, dispersão, barras, área. Porém neste pequeno artigo vamos abordar o gráfico de waffle.

### **O que é um Gráfico de Waffle?**

Ele é conceitualmente bem simples. Basicamente, um gráfico de Waffle, também chamado de gráfico de torta quadrada, é uma forma de visualização de dados que podemos usar quando queremos representar a distribuição proporcional de um conjunto de categorias. É bem parecido com um gráfico de pizza tradicional, só que no lugar das fatias, usamos quadrados de mesmo tamanho.

O gráfico de Waffle é útil quando desejamos apresentar várias categorias de dados em uma única imagem, de modo a facilitar a comparação. É especialmente adequado para dados discretos e que podem ser divididos em partes iguais, como preferência de marcas, porcentagens de um todo, entre outros.

Neste artigo, veremos as etapas necessárias para criação de um gráfico de waffle usando Python.




### **Plotando um Gráfico de Waffle**

Existem diversas ferramentas de visualizaçao de dados, neste exercício vamos utilizar uma das mais populares e importantes bibliotecas em Python, denominada [Matplotlib](https://matplotlib.org/), que funciona muito bem para gráficos de barras, de dispersão, de linhas, plotar histogramas e gráficos de waffle. Trata-se de uma biblioteca bastante ampla para criar visualizações estáticas, animadas e interativas. Também utilizaremos a biblioteca [pywaffle](https://pywaffle.readthedocs.io/en/latest/) para criar o gráfico de waffle e a biblioteca [pandas](https://pandas.pydata.org/docs/) para criarmos o dataframe de exemplo.

Vejamos como é simples e rápido criar um gráfico de waffle no python. 

Neste exemplo, vamos imaginar que temos um conjunto de dados que contém a preferência de marcas de cerveja entre um grupo de amigos, conforme a tabela abaixo:

```python
import pandas as pd

# criar DataFrame usando dicionário
dicionário = {
    'Marca': ['Brahma', 'Skol', 'Bohemia', 'Heineken', 'Stella Artois'],
    'Quantidade': [15, 10, 5, 20, 8]
}

dados_df = pd.DataFrame(dicionário)

# exibir o DataFrame
display(dados_df)
```

Para criar um gráfico de Waffle, precisamos das seguintes bibliotecas:

Matplotlib: para criar o gráfico

Pywaffle: para criar o gráfico de Waffle

Como não estamos falando de python nativo, precisamos antes instalar o pywaffle. E em seguida as bibliotecas necessárias.

```python
# instalar o pywaffle
!pip install pywaffle --q
```

```python
# importando as bibliotecas necessárias
from matplotlib import pyplot as plt
from pywaffle import Waffle
```

Agora precisamos preparar nossos dados. Vamos criar um dicionário com as informações sobre as marcas de cerveja e a quantidade que cada uma foi escolhida.

```python
# definindo as cores do gráfico
cores = ['#f1c40f', '#9b59b6', '#3498db', 
         '#2ecc71', '#e74c3c']

# Cria o gráfico de waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=8,
                columns=14,
                colors=(cores),
                values={'Brahma': 7, 
                        'Bohemia': 5,
                        'Skol': 10,
                        'Heinenken': 25,
                        'Stella Artois': 16
                        },
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1,1)},
                 icons='wine-bottle', icon_size=24,
                 icon_legend=True
            )

# Define o título do gráfico

fig.suptitle('Preferência de Marcas de Cerveja', fontsize=18)
fig.set_tight_layout(False)

# Mostra o gráfico
plt.show()
```
